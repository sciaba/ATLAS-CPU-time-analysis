```python
import json
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
input_file = 'cpudetails_HS06_CERN.json'
with open(input_file) as f:
    data = json.load(f)
```

```python
results = [d['value'] for d in data['rows'] if d['value'].has_key('HEPSPEC') and d['value']['HEPSPEC'] != 'None' and d['value']['HEPSPEC'] != '1.00']
```

```python
df = DataFrame.from_dict(results)
```

```python
df
```

```python
set(df['PROCESSOR MODEL'])
```

```python
df['HEPSPEC'] = df['HEPSPEC'].astype(np.float)
df['NUM OF LOGICAL CORES'] = df['NUM OF LOGICAL CORES'].astype(np.int)
df['NUM OF PROCESSORS'] = df['NUM OF PROCESSORS'].astype(np.int)
df['NUM OF PHYSICAL CORES'] = df['NUM OF PHYSICAL CORES'].astype(np.int)
df['HS06_core'] = df['HEPSPEC'].divide(df['NUM OF LOGICAL CORES'])
df['Cores_CPU'] = df['NUM OF LOGICAL CORES'].divide(df['NUM OF PROCESSORS'])
```
